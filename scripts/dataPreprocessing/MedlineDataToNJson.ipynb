```python
import pubmed_parser as pp
import json
```

```python

dicts_data = pp.parse_medline_xml('../data/pubmed23n0001.xml.gz',
                                 year_info_only=False,
                                 nlm_category=False,
                                 author_list=True,
                                 reference_list=False) 
```

```python
# return list of dictionaries 
dicts_data
```

```python
#Export dicstionariy to 1.json file
with open("1.json", "w") as outfile:
    json.dump(dicts_data, outfile)
```

```python
# Make 1.json file a new-line delimited collection of dictionaries in the list
with open("1.json", 'w') as f:
    for d in dicts_data:
        json.dump(d, f)
        f.write('\n')
```
